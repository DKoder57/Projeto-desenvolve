Primeiramente criaremos a classe abstrata `EventoABC` com os métodos de instância abstratos `__str__(self)` e `isConcluido(self)`, indicando que todos as subclasses que dela herdarem devem implementar esses métodos.

`EventoABC` também possui os atributos `_titulo` (string) e `_descricao` (string), cujos valores são recebidos e inicializados no construtor da classe. Note a convenção de nomenclatura indicando o caráter privado desses atributos.

````python
#### Classe EventoABC

from abc import ABC, abstractmethod

class EventoABC(ABC):
    def __init__(self, titulo, descricao):
        self._titulo = titulo
        self._descricao = descricao

    @abstractmethod
    def __str__(self): pass

    @abstractmethod
    def isConcluido(self): pass
````

````python
from datetime import datetime, timedelta

class B:
    FORMAT = '%d/%m/%Y, %H:%M'

    @property
    def x(self):
        return self._x.strftime(self.FORMAT)

    @x.setter
    def x(self, v):
        try:
            self._x = datetime.strptime(v, self.FORMAT)
        except:
            raise ValueError

    def p(self):
        return self._x < datetime.now()

    def s(self, n):
        return (self._x + timedelta(days=n)).strftime(self.FORMAT)


b = B()
b.x = '05/02/2024, 12:30'
b.x = b.s(30)
print(b.x, b.p())
````

````python
class EventoLista(A):
    def __init__(self, t, d, i, f, k):
        super().__init__(t, d)
        self._l = []

        tmp = B()
        tmp.x = i

        while True:
            n = B()
            n.x = tmp.x
            self._l.append(n)
            if tmp.x >= f: break
            tmp.x = tmp.s(k)

    def isConcluido(self, i):
        return self._l[i].p()

    def __str__(self):
        r = ""
        for i, v in enumerate(self._l):
            r += f"{self._t} {v.x} {self._d} {self.isConcluido(i)}\n"
        return r

    def editar(self, a, n):
        for v in self._l:
            if v.x == a:
                v.x = n


el = EventoLista('Reunião', 'Sala', '05/01/2024, 16:30', '05/03/2024, 16:30', 30)
print(el)
el.editar('05/02/2024, 16:30', '05/02/2024, 10:30')
print(el)
````
